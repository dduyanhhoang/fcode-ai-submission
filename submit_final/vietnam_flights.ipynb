```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from datetime import datetime, timedelta
from itertools import combinations
from loguru import logger

WAIT_TIMEOUT = 10

# Setup loguru
logger.remove()  # Remove default handler
logger.add("scraping.log", rotation="500 MB", level="INFO")
logger.add(lambda msg: print(msg, end=""), colorize=True, level="INFO")

# Setup
chrome_options = Options()
chrome_options.add_argument("--incognito")
driver = uc.Chrome(options=chrome_options)

# Generate URLs
airports = ['HAN', 'DAD', 'PQC', 'CXR', 'DLI', 'SGN']
classes = ['ECONOMY', 'PREMIUM_ECONOMY', 'BUSINESS']
start_date = datetime(2024, 8, 8)
end_date = datetime(2024, 12, 31)
base_url = "https://www.traveloka.com/en-vn/flight/fullsearch?ap={}.{}&dt={}.NA&ps=1.0.0&sc={}"

urls = []
for dep, arr in combinations(airports, 2):
    for travel_class in classes:
        current_date = start_date
        while current_date <= end_date:
            date_str = current_date.strftime("%d-%m-%Y")
            urls.append(base_url.format(dep, arr, date_str, travel_class))
            urls.append(base_url.format(arr, dep, date_str, travel_class))
            current_date += timedelta(days=1)

total_flights = 0

def get_flights(driver):
    flight_xpaths = [
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[5]/div[3]/div/div/div[@data-testid]',
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[6]/div[3]/div/div/div/div[@data-testid]'
    ]
    
    for xpath in flight_xpaths:
        flights = driver.find_elements(By.XPATH, xpath)
        valid_flights = [flight for flight in flights if flight.get_attribute('data-testid').startswith('flight-inventory-card-container-')]
        if valid_flights:
            return valid_flights
    return []

for url in urls:
    logger.info(f"Processing URL: {url}")
    driver.get(url)
    
    try:
        # Wait for the flight search result content to be present
        WebDriverWait(driver, WAIT_TIMEOUT).until(
            EC.presence_of_element_located((By.ID, "FLIGHT_SEARCH_RESULT_CONTENT"))
        )

        # Scroll to the bottom until no more content loads
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Reduced wait time for scrolling
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        # Find all flight divs
        valid_flights = get_flights(driver)
        
        num_flights = len(valid_flights)
        total_flights += num_flights

        logger.success(f"URL: {url}")
        logger.success(f"Number of flights: {num_flights}")
        logger.info("---")

    except Exception as e:
        logger.error(f"Error processing URL {url}: {str(e)}")

    # Add a small delay between requests
    time.sleep(2)

logger.success(f"Total number of flights across all URLs: {total_flights}")

driver.quit()
```

```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from datetime import datetime
from loguru import logger
from bs4 import BeautifulSoup
import pandas as pd

WAIT_TIMEOUT = 10

# Setup loguru
logger.remove()  # Remove default handler
logger.add("scraping.log", rotation="500 MB", level="INFO")
logger.add(lambda msg: print(msg, end=""), colorize=True, level="INFO")

# Setup
chrome_options = Options()
chrome_options.add_argument("--incognito")
driver = uc.Chrome(options=chrome_options)

# Test URL
test_url = "https://www.traveloka.com/en-vn/flight/fullsearch?ap=SGN.HAN&dt=8-8-2024.NA&ps=1.0.0&sc=ECONOMY"

def get_flights(driver):
    flight_xpaths = [
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[5]/div[3]/div/div/div[@data-testid]',
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[6]/div[3]/div/div/div/div[@data-testid]'
    ]
    
    for xpath in flight_xpaths:
        flights = driver.find_elements(By.XPATH, xpath)
        valid_flights = [flight for flight in flights if flight.get_attribute('data-testid').startswith('flight-inventory-card-container-')]
        if valid_flights:
            return valid_flights
    return []

def extract_flight_details(flight):
    buttons_xpath = './/div/div/div/div/div[2]/div[1]/div'
    buttons = flight.find_elements(By.XPATH, buttons_xpath)
    
    if len(buttons) >= 5:
        # Flight Details
        driver.execute_script("arguments[0].click();", buttons[0])
        time.sleep(1)
        details_area = flight.find_element(By.XPATH, './/div/div/div/div/div[3]/div/div/div/div')
        details_html = BeautifulSoup(details_area.get_attribute('innerHTML'), 'html.parser')
        
        details_info = details_html.find_all('div', {'class': 'css-1dbjc4n r-1tuna9m r-1oszu61 r-18u37iz r-1h0z5md'})
        
        dep_time = details_info[0].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-5oul0u r-fdjqy7'}).text
        des_time = details_info[2].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7'}).text
        
        duration = details_info[1].find('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1enofrn r-majxgm r-1cwl3u0 r-fdjqy7'}).text
        hours, minutes = 0, 0
        if 'h' in duration:
            hours = int(duration.split('h')[0].strip())
        if 'm' in duration:
            minutes = int(duration.split('h')[1].strip().replace('m', ''))
        duration = round(hours + minutes / 60.0, 2)
        
        aircraft = details_info[1].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'}).text
        airline = details_info[1].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7'}).text
        
        baggage = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})[0].text
        c_baggage = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})[1].text
        
        mid_info = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})
        entertainment = 1 if 'entertainment' in str(mid_info).lower() else 0
        meal = 1 if 'meal' in str(mid_info).lower() else 0
        power_supply = 1 if 'usb port' in str(mid_info).lower() else 0
        
        driver.execute_script("arguments[0].click();", buttons[0])
        time.sleep(0.5)
        
        # Additional Benefits
        driver.execute_script("arguments[0].click();", buttons[1])
        time.sleep(1)
        benefits_area = flight.find_element(By.XPATH, './/div/div/div/div/div[3]/div/div/div/div')
        benefits_html = BeautifulSoup(benefits_area.get_attribute('innerHTML'), 'html.parser')
        
        refundable = 1 if 'non-refundable' not in benefits_html.text.lower() and 'refund info unavailable' not in benefits_html.text.lower() else 0
        reschedulable = 1 if 'reschedule available' in benefits_html.text.lower() else 0
        
        driver.execute_script("arguments[0].click();", buttons[1])
        time.sleep(0.5)
        
        # Promos
        driver.execute_script("arguments[0].click();", buttons[-1])
        time.sleep(1)
        promos_area = flight.find_element(By.XPATH, './/div/div/div/div/div[3]/div/div/div/div')
        promos_html = BeautifulSoup(promos_area.get_attribute('innerHTML'), 'html.parser')
        
        promos = len(promos_html.find_all('div', {'class': 'css-1dbjc4n r-14lw9ot r-kdyh1x r-5m9fi9 r-1pi2tsx r-1rd2zbf r-ggk5by r-xyw6el'}))
        
        driver.execute_script("arguments[0].click();", buttons[-1])
        time.sleep(0.5)
        
        return {
            'dep_time': dep_time,
            'des_time': des_time,
            'duration': duration,
            'aircraft': aircraft,
            'airline': airline,
            'baggage': baggage,
            'c_baggage': c_baggage,
            'entertainment': entertainment,
            'meal': meal,
            'power_supply': power_supply,
            'refundable': refundable,
            'reschedulable': reschedulable,
            'promos': promos
        }
    else:
        logger.warning(f"Expected at least 5 buttons, but found {len(buttons)} for a flight")
        return None

logger.info(f"Processing URL: {test_url}")
driver.get(test_url)

try:
    WebDriverWait(driver, WAIT_TIMEOUT).until(
        EC.presence_of_element_located((By.ID, "FLIGHT_SEARCH_RESULT_CONTENT"))
    )

    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    valid_flights = get_flights(driver)
    
    if valid_flights:
        first_flight = valid_flights[0]
        flight_info = extract_flight_details(first_flight)
        
        if flight_info:
            logger.success("Successfully extracted flight details:")
            for key, value in flight_info.items():
                logger.info(f"{key}: {value}")
        else:
            logger.error("Failed to extract flight details")
    else:
        logger.error("No valid flights found")

except Exception as e:
    logger.error(f"Error processing URL {test_url}: {str(e)}")

driver.quit()
```

```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from datetime import datetime
from loguru import logger
from bs4 import BeautifulSoup
import pandas as pd

WAIT_TIMEOUT = 10

# Setup loguru
logger.remove()  # Remove default handler
logger.add("scraping.log", rotation="500 MB", level="INFO")
logger.add(lambda msg: print(msg, end=""), colorize=True, level="INFO")

# Setup
chrome_options = Options()
chrome_options.add_argument("--incognito")
driver = uc.Chrome(options=chrome_options)

# Test URL
test_url = "https://www.traveloka.com/en-vn/flight/fullsearch?ap=SGN.HAN&dt=8-8-2024.NA&ps=1.0.0&sc=ECONOMY"

def get_flights(driver):
    flight_xpaths = [
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[5]/div[3]/div/div/div[@data-testid]',
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[6]/div[3]/div/div/div/div[@data-testid]'
    ]
    
    for xpath in flight_xpaths:
        flights = driver.find_elements(By.XPATH, xpath)
        valid_flights = [flight for flight in flights if flight.get_attribute('data-testid').startswith('flight-inventory-card-container-')]
        if valid_flights:
            return valid_flights
    return []

def extract_flight_details(flight):
    buttons_xpath = './/div/div/div/div/div[2]/div[1]/div'
    buttons = flight.find_elements(By.XPATH, buttons_xpath)
    
    if len(buttons) >= 5:
        # Flight Details
        driver.execute_script("arguments[0].click();", buttons[0])
        time.sleep(1)
        details_area = flight.find_element(By.XPATH, './/div/div/div/div/div[3]/div/div/div/div')
        details_html = BeautifulSoup(details_area.get_attribute('innerHTML'), 'html.parser')
        
        details_info = details_html.find_all('div', {'class': 'css-1dbjc4n r-1tuna9m r-1oszu61 r-18u37iz r-1h0z5md'})
        
        dep_time = details_info[0].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-5oul0u r-fdjqy7'}).text
        des_time = details_info[2].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7'}).text
        
        duration = details_info[1].find('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1enofrn r-majxgm r-1cwl3u0 r-fdjqy7'}).text
        hours, minutes = 0, 0
        if 'h' in duration:
            hours = int(duration.split('h')[0].strip())
        if 'm' in duration:
            minutes = int(duration.split('h')[1].strip().replace('m', ''))
        duration = round(hours + minutes / 60.0, 2)
        
        aircraft = details_info[1].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'}).text
        airline = details_info[1].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7'}).text
        
        baggage = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})[0].text
        c_baggage = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})[1].text
        
        mid_info = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})
        entertainment = 1 if 'entertainment' in str(mid_info).lower() else 0
        meal = 1 if 'meal' in str(mid_info).lower() else 0
        power_supply = 1 if 'usb port' in str(mid_info).lower() else 0
        
        driver.execute_script("arguments[0].click();", buttons[0])
        time.sleep(0.5)
        
        # Additional Benefits
        driver.execute_script("arguments[0].click();", buttons[1])
        time.sleep(1)
        benefits_area = flight.find_element(By.XPATH, './/div/div/div/div/div[3]/div/div/div/div')
        benefits_html = BeautifulSoup(benefits_area.get_attribute('innerHTML'), 'html.parser')
        
        refundable = 1 if 'non-refundable' not in benefits_html.text.lower() and 'refund info unavailable' not in benefits_html.text.lower() else 0
        reschedulable = 1 if 'reschedule available' in benefits_html.text.lower() else 0
        
        driver.execute_script("arguments[0].click();", buttons[1])
        time.sleep(0.5)
        
        # Promos
        driver.execute_script("arguments[0].click();", buttons[-1])
        time.sleep(1)
        promos_area = flight.find_element(By.XPATH, './/div/div/div/div/div[3]/div/div/div/div')
        promos_html = BeautifulSoup(promos_area.get_attribute('innerHTML'), 'html.parser')
        
        promos = len(promos_html.find_all('div', {'class': 'css-1dbjc4n r-14lw9ot r-kdyh1x r-5m9fi9 r-1pi2tsx r-1rd2zbf r-ggk5by r-xyw6el'}))
        
        driver.execute_script("arguments[0].click();", buttons[-1])
        time.sleep(0.5)
        
        return {
            'dep_time': dep_time,
            'des_time': des_time,
            'duration': duration,
            'aircraft': aircraft,
            'airline': airline,
            'baggage': baggage,
            'c_baggage': c_baggage,
            'entertainment': entertainment,
            'meal': meal,
            'power_supply': power_supply,
            'refundable': refundable,
            'reschedulable': reschedulable,
            'promos': promos
        }
    else:
        logger.warning(f"Expected at least 5 buttons, but found {len(buttons)} for a flight")
        return None

logger.info(f"Processing URL: {test_url}")
driver.get(test_url)

try:
    WebDriverWait(driver, WAIT_TIMEOUT).until(
        EC.presence_of_element_located((By.ID, "FLIGHT_SEARCH_RESULT_CONTENT"))
    )

    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    valid_flights = get_flights(driver)
    
    if valid_flights:
        first_flight = valid_flights[0]
        flight_info = extract_flight_details(first_flight)
        
        if flight_info:
            logger.success("Successfully extracted flight details:")
            for key, value in flight_info.items():
                logger.info(f"{key}: {value}")
            
            # Save to DataFrame
            df = pd.DataFrame([flight_info])
            logger.success("Flight data saved to DataFrame")
            logger.info(f"DataFrame shape: {df.shape}")
            logger.info(f"DataFrame columns: {df.columns.tolist()}")
            
            # You can perform further operations on the DataFrame here
            # For example:
            # df['new_column'] = some_calculation(df)
            # df = df[df['some_column'] > some_value]
            # etc.
            
        else:
            logger.error("Failed to extract flight details")
    else:
        logger.error("No valid flights found")

except Exception as e:
    logger.error(f"Error processing URL {test_url}: {str(e)}")

driver.quit()
```

```python
from selenium import webdriver
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, StaleElementReferenceException
import time
from datetime import datetime, timedelta
from itertools import combinations
from loguru import logger
from bs4 import BeautifulSoup
import pandas as pd

WAIT_TIMEOUT = 20  # Increased timeout

# Setup loguru
logger.remove()  # Remove default handler
logger.add("scraping.log", rotation="500 MB", level="INFO")
logger.add(lambda msg: print(msg, end=""), colorize=True, level="INFO")

# Setup
chrome_options = Options()
chrome_options.add_argument("--incognito")
driver = uc.Chrome(options=chrome_options)

# Generate URLs
airports = ['HAN', 'DAD', 'PQC', 'CXR', 'DLI', 'SGN']
classes = ['ECONOMY', 'PREMIUM_ECONOMY', 'BUSINESS']
start_date = datetime(2024, 8, 8)
end_date = datetime(2024, 12, 31)
base_url = "https://www.traveloka.com/en-vn/flight/fullsearch?ap={}.{}&dt={}.NA&ps=1.0.0&sc={}"

urls = []
for dep, arr in combinations(airports, 2):
    for travel_class in classes:
        current_date = start_date
        while current_date <= end_date:
            date_str = current_date.strftime("%d-%m-%Y")
            urls.append(base_url.format(dep, arr, date_str, travel_class))
            urls.append(base_url.format(arr, dep, date_str, travel_class))
            current_date += timedelta(days=1)

def get_flights(driver):
    flight_xpaths = [
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[5]/div[3]/div/div/div[@data-testid]',
        '//*[@id="FLIGHT_SEARCH_RESULT_CONTENT"]/div[6]/div[3]/div/div/div/div[@data-testid]'
    ]
    
    for xpath in flight_xpaths:
        flights = driver.find_elements(By.XPATH, xpath)
        valid_flights = [flight for flight in flights if flight.get_attribute('data-testid').startswith('flight-inventory-card-container-')]
        if valid_flights:
            return valid_flights
    return []

def extract_flight_details(flight):
    try:
        buttons_xpath = './/div/div/div/div/div[2]/div[1]/div'
        buttons = flight.find_elements(By.XPATH, buttons_xpath)
        
        if len(buttons) < 3:
            logger.warning(f"Expected at least 3 buttons, but found {len(buttons)} for a flight")
            return None

        # Flight Details
        driver.execute_script("arguments[0].click();", buttons[0])
        time.sleep(1)
        details_area = WebDriverWait(flight, WAIT_TIMEOUT).until(
            EC.presence_of_element_located((By.XPATH, './/div/div/div/div/div[3]/div/div/div/div'))
        )
        details_html = BeautifulSoup(details_area.get_attribute('innerHTML'), 'html.parser')
        
        details_info = details_html.find_all('div', {'class': 'css-1dbjc4n r-1tuna9m r-1oszu61 r-18u37iz r-1h0z5md'})
        
        if len(details_info) < 3:
            logger.warning(f"Not enough details found. Found {len(details_info)} elements.")
            return None
        
        dep_time = details_info[0].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-5oul0u r-fdjqy7'})
        des_time = details_info[2].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7'})
        
        if not dep_time or not des_time:
            logger.warning("Could not find departure or arrival time.")
            return None
        
        dep_time = dep_time.text
        des_time = des_time.text
        
        duration_element = details_info[1].find('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1enofrn r-majxgm r-1cwl3u0 r-fdjqy7'})
        if not duration_element:
            logger.warning("Could not find duration element.")
            return None
        
        duration = duration_element.text
        hours, minutes = 0, 0
        if 'h' in duration:
            hours = int(duration.split('h')[0].strip())
        if 'm' in duration:
            minutes = int(duration.split('h')[1].strip().replace('m', ''))
        duration = round(hours + minutes / 60.0, 2)
        
        aircraft = details_info[1].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})
        airline = details_info[1].find('div', {'class': 'css-901oao r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-fdjqy7'})
        
        if not aircraft or not airline:
            logger.warning("Could not find aircraft or airline information.")
            return None
        
        aircraft = aircraft.text
        airline = airline.text
        
        baggage_elements = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})
        if len(baggage_elements) < 2:
            logger.warning("Could not find baggage information.")
            return None
        
        baggage = baggage_elements[0].text
        c_baggage = baggage_elements[1].text
        
        mid_info = details_info[1].find_all('div', {'class': 'css-901oao r-13awgt0 r-t1w4ow r-1b43r93 r-majxgm r-rjixqe r-19u6a5r r-fdjqy7'})
        entertainment = 1 if 'entertainment' in str(mid_info).lower() else 0
        meal = 1 if 'meal' in str(mid_info).lower() else 0
        power_supply = 1 if 'usb port' in str(mid_info).lower() else 0
        
        driver.execute_script("arguments[0].click();", buttons[0])
        time.sleep(0.5)
        
        # Additional Benefits
        driver.execute_script("arguments[0].click();", buttons[1])
        time.sleep(1)
        benefits_area = WebDriverWait(flight, WAIT_TIMEOUT).until(
            EC.presence_of_element_located((By.XPATH, './/div/div/div/div/div[3]/div/div/div/div'))
        )
        benefits_html = BeautifulSoup(benefits_area.get_attribute('innerHTML'), 'html.parser')
        
        refundable = 1 if 'non-refundable' not in benefits_html.text.lower() and 'refund info unavailable' not in benefits_html.text.lower() else 0
        reschedulable = 1 if 'reschedule available' in benefits_html.text.lower() else 0
        
        driver.execute_script("arguments[0].click();", buttons[1])
        time.sleep(0.5)
        
        # Promos
        driver.execute_script("arguments[0].click();", buttons[-1])
        time.sleep(1)
        promos_area = WebDriverWait(flight, WAIT_TIMEOUT).until(
            EC.presence_of_element_located((By.XPATH, './/div/div/div/div/div[3]/div/div/div/div'))
        )
        promos_html = BeautifulSoup(promos_area.get_attribute('innerHTML'), 'html.parser')
        
        promos = len(promos_html.find_all('div', {'class': 'css-1dbjc4n r-14lw9ot r-kdyh1x r-5m9fi9 r-1pi2tsx r-1rd2zbf r-ggk5by r-xyw6el'}))
        
        driver.execute_script("arguments[0].click();", buttons[-1])
        time.sleep(0.5)
        
        return {
            'dep_time': dep_time,
            'des_time': des_time,
            'duration': duration,
            'aircraft': aircraft,
            'airline': airline,
            'baggage': baggage,
            'c_baggage': c_baggage,
            'entertainment': entertainment,
            'meal': meal,
            'power_supply': power_supply,
            'refundable': refundable,
            'reschedulable': reschedulable,
            'promos': promos
        }
    except (NoSuchElementException, TimeoutException, StaleElementReferenceException) as e:
        logger.error(f"Error extracting flight details: {str(e)}")
        return None
    except Exception as e:
        logger.error(f"Unexpected error extracting flight details: {str(e)}")
        return None

all_flights_data = []

for url in urls:
    logger.info(f"Processing URL: {url}")
    driver.get(url)
    
    try:
        WebDriverWait(driver, WAIT_TIMEOUT).until(
            EC.presence_of_element_located((By.ID, "FLIGHT_SEARCH_RESULT_CONTENT"))
        )

        # Improved scrolling mechanism
        scroll_attempts = 0
        last_height = driver.execute_script("return document.body.scrollHeight")
        while scroll_attempts < 3:  # Limit scroll attempts
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                scroll_attempts += 1
            else:
                scroll_attempts = 0
                last_height = new_height

        valid_flights = get_flights(driver)
        
        for flight in valid_flights:
            flight_info = extract_flight_details(flight)
            if flight_info:
                flight_info['url'] = url
                all_flights_data.append(flight_info)
        
        logger.success(f"Processed {len(valid_flights)} flights for URL: {url}")

    except Exception as e:
        logger.error(f"Error processing URL {url}: {str(e)}")

    time.sleep(2)

# Create DataFrame from all collected flight data
df = pd.DataFrame(all_flights_data)

logger.success("All flight data saved to DataFrame")
logger.info(f"Total flights processed: {len(df)}")
logger.info(f"DataFrame shape: {df.shape}")
logger.info(f"DataFrame columns: {df.columns.tolist()}")

driver.quit()

# You can perform further operations on the DataFrame here
# For example:
# df['new_column'] = some_calculation(df)
# df = df[df['some_column'] > some_value]
# etc.
```
