Follow the description from the [data set description](https://www.kaggle.com/datasets/rashikrahmanpritom/heart-attack-analysis-prediction-dataset/data), these are some explanations of the attributes from the data set.  
- **Age**: Age of the patient
- **Sex**: Sex of the patient
- **exng**: exercise induced angina (1 = yes; 0 = no)
- **ca**: number of major vessels (0-3)
- **cp**: Chest Pain type 
    - Value 1: typical angina
    - Value 2: atypical angina
    - Value 3: non-anginal pain
    - Value 4: asymptomatic
- **trtbps**: resting blood pressure (in mm Hg)
- **chol**: cholestoral in mg/dl fetched via BMI sensor
- **fbs**: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
- **rest_ecg**: resting electrocardiographic results
    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
- **thalach**: maximum heart rate achieved
- **target**: 0= less chance of heart attack 1= more chance of heart attack

# Step 2: Univariate analysis
There are categorical and continuous data in the dataset, here I split them to do some **univariate analysis**

```python
categorical_features = ['sex','exng','caa','cp','fbs','restecg','slp','thall']
continuous_features = ["age", "trtbps", "chol", "thalachh", "oldpeak"]
target_col = ["output"]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Count plot of 'output'
sns.countplot(x='output', data=df, ax=axes[0])
axes[0].set_title('Count Plot of Output')
axes[0].set_xlabel('Output')
axes[0].set_ylabel('Count')

# Pie chart of 'output'
df['output'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=axes[1])
axes[1].set_title('Pie Chart of Output')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# Calculate the number of rows needed
num_rows = (len(categorical_features) + 1) // 2

# Create subplots
fig, axes = plt.subplots(num_rows, 4, figsize=(20, 5 * num_rows))

# Flatten the axes array for easy iteration
axes = axes.flatten()

for i, feature in enumerate(categorical_features):
    # Bar plot (countplot)
    sns.countplot(x=feature, data=df, ax=axes[2 * i])
    axes[2 * i].set_title(f'Bar Plot of {feature}')
    axes[2 * i].set_xlabel(feature)
    axes[2 * i].set_ylabel('Count')
    
    # Pie chart
    value_counts = df[feature].value_counts()
    value_counts.plot.pie(autopct='%1.1f%%', startangle=90, ax=axes[2 * i + 1])
    axes[2 * i + 1].set_title(f'Pie Chart of {feature}')
    axes[2 * i + 1].set_ylabel('')

# Hide any unused subplots
for j in range(2 * len(categorical_features), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axes = plt.subplots(len(continuous_features), 3, figsize=(15, 5 * len(continuous_features)))

for i, feature in enumerate(continuous_features):
    # Descriptive statistics
    desc_stats = df[feature].describe()
    axes[i, 0].text(0.5, 0.5, str(desc_stats), fontsize=12, verticalalignment='center', horizontalalignment='center')
    axes[i, 0].set_title(f'Descriptive Stats of {feature}')
    axes[i, 0].axis('off')
    
    # Histogram
    df[feature].hist(bins=30, edgecolor='k', ax=axes[i, 1])
    axes[i, 1].set_title(f'Histogram of {feature}')
    axes[i, 1].set_xlabel(feature)
    axes[i, 1].set_ylabel('Frequency')
    
    # Box plot
    sns.boxplot(x=df[feature], ax=axes[i, 2])
    axes[i, 2].set_title(f'Box Plot of {feature}')
    axes[i, 2].set_xlabel(feature)

plt.tight_layout()
plt.show()
```

# Step 3: Bivariate analysis
Perform bivariate analysis to explore relationships between pairs of variables

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='chol', hue='output', data=df)
plt.title('Scatter Plot of Age vs Cholesterol')
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='cp', y='thalachh', data=df)
plt.title('Box Plot of Chest Pain Type vs Maximum Heart Rate')
plt.show()

# Heatmap for correlation matrix
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()

# Pair plot
sns.pairplot(df, hue='output')
plt.show()
```

```python
# Create subplots with 3 plots per line
rows = (len(continuous_features) + 2) // 3  # Calculate the number of rows needed
fig, axes = plt.subplots(rows, 3, figsize=(18, 6 * rows))

for i, feature in enumerate(continuous_features):
    row, col = divmod(i, 3)
    sns.boxplot(x='output', y=feature, data=df, ax=axes[row, col])
    axes[row, col].set_title(f'Box Plot of {feature} vs Output')
    axes[row, col].set_xlabel('Output')
    axes[row, col].set_ylabel(feature)

# Remove any empty subplots
for j in range(i + 1, rows * 3):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()
```

```python
# Create subplots with 3 plots per line
rows = (len(categorical_features) + 2) // 3  # Calculate the number of rows needed
fig, axes = plt.subplots(rows, 3, figsize=(18, 6 * rows))

for i, feature in enumerate(categorical_features):
    row, col = divmod(i, 3)
    sns.countplot(x=feature, hue='output', data=df, ax=axes[row, col])
    axes[row, col].set_title(f'Bar Plot of {feature} vs Output')
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('Count')

# Remove any empty subplots
for j in range(i + 1, rows * 3):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()
```
